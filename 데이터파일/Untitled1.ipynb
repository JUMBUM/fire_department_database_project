```python
관할구역_df = pd.read_csv("jurisdiction.csv", encoding='cp949')

관할구역_firestation_id = list(관할구역_df['firestation_id'])
관할구역_region_id = list(관할구역_df['region_id'])

관할구역_df.head(3)
```

```python
소방서현황_df = pd.read_csv("status_firestation_sejong.csv", encoding='cp949')

소방서현황_firestation_id = list(소방서현황_df['firestation_id'])
소방서현황_team_num = list(소방서현황_df['team_num'])

소방서현황_df.head(3)
```

```python
신고_report_id = []
신고_firestation_id = []
신고_team_num = []
사건_region_id = []
사건_case_id = []
사건_case_type = []
types = ['F', 'R', 'A']


관할구역_dict = {}
for firestation_id in 소방서현황_firestation_id:
    firestation_id.strip()
    i_ls = []
    for i in range(len(관할구역_firestation_id)):
        if firestation_id == 관할구역_firestation_id[i].strip():
            i_ls.append(i)

    지역__ = []
    for ii in i_ls:
        지역__.append(관할구역_region_id[ii])
    
    관할구역_dict.update({firestation_id:지역__})

관할구역_dict_소방서 = list(관할구역_dict.keys())
관할구역_dict_지역 = list(관할구역_dict.values())



for i in range(1,301):
    print(i)
    x = i
    if x < 10:
        x = "00"+str(x)
    elif x < 100:
        x = "0"+str(x)
    신고_report_id.append("ER{}".format(x))
    사건_case_id.append("IC{}".format(x))
    사건_case_type.append(np.random.choice(types))
    
    소방서 = np.random.choice(소방서현황_firestation_id)
    신고_firestation_id.append(소방서)
    신고_team_num.append(np.random.randint(1, 1+int(소방서현황_team_num[소방서현황_firestation_id.index(소방서)])))
    print(소방서, i)
    사건_region_id.append(np.random.choice(관할구역_dict_지역[관할구역_dict_소방서.index(소방서)]))
    
```
